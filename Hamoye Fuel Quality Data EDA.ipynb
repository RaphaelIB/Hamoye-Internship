```python
#returns the correlation values of the features in the dataset
fuel_ds.corr()
```

```python
#For the fuel type coal, what is the percentage change in the fuel cost per unit burned in 1998 compared to 1994?
coal_fuel_type = fuel_ds[fuel_ds['fuel_type_code_pudl']=='coal']
burnt_in_1994 = coal_fuel_type[coal_fuel_type['report_year']==1994]
burnt_in_1998 = coal_fuel_type[coal_fuel_type['report_year']==1998]
print(burnt_in_1998['fuel_cost_per_unit_burned'].sum())
print(burnt_in_1994['fuel_cost_per_unit_burned'].sum())
print()
```

```python
#Which year has the highest average fuel cost per unit delivered?
year = fuel_ds.groupby('report_year')
year['fuel_cost_per_unit_delivered'].mean().sort_values()
```

```python
fuel_type_code = fuel_ds.groupby('fuel_type_code_pudl')
```

```python
#Which of the following fuel type code has the lowest average fuel cost per unit burned?
fuel_type_code = fuel_ds.groupby('fuel_type_code_pudl')
fuel_type_code['fuel_cost_per_unit_burned'].mean().sort_values()
```
